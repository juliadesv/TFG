```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
sns.factorplot(x='Time_segment',y='Density', data=df, palette='Set1')
plt.title('Number of accidents sorted by month')
plt.xlabel('Month', fontsize=10)
plt.ylabel('Number of accidents', fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python
 import matplotlib.pyplot as plt

# Suponiendo que 'data' es tu DataFrame y 'tiempo' es una columna de fechas/tiempo
# 'variable_numerica' es la variable numérica que deseas graficar en función del tiempo

plt.figure(figsize=(10, 6))  # Definir el tamaño del gráfico (opcional)

plt.plot(df['Day'], df['TM'], marker='o', linestyle='-')

plt.xlabel('Tiempo')  # Etiqueta del eje X
plt.ylabel('Variable Numérica')  # Etiqueta del eje Y
plt.title('Variable Numérica en Función del Tiempo')  # Título del gráfico

plt.tight_layout()  # Ajustar el diseño del gráfico

plt.show()  # Mostrar el gráfico
```

```python
sns.factorplot(x='TM', data=df, kind='count', palette='Set1')
plt.title('Number of accidents sorted by month')
plt.xlabel('Month', fontsize=10)
plt.ylabel('Number of accidents', fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python
df['PPT'] = df['PPT'].str.replace(',', '.').astype(float)

# Mostrar el DataFrame resultante
print(df)
```

```python
average_temperature = df[df['TM'].notnull()].groupby('Time_segment')['TM'].mean()

# Crear el gráfico
plt.figure(figsize=(10, 6))  # Definir el tamaño del gráfico (opcional)

# Graficar el promedio de temperatura por segmento de tiempo
average_temperature.plot(kind='bar', color='skyblue')

plt.xlabel('Segmento de Tiempo')  # Etiqueta del eje X
plt.ylabel('Temperatura Promedio')  # Etiqueta del eje Y
plt.title('Temperatura Promedio por Segmento de Tiempo')  # Título del gráfico

plt.xticks(rotation=45)  # Rotar las etiquetas del eje X para mejor legibilidad

plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()  # Mostrar el gráfico
```

```python
import matplotlib.pyplot as plt

# Suponiendo que 'data' es tu DataFrame
# 'TM' es la columna de temperatura
# 'time_segment' es la columna que define los segmentos de tiempo

# Calcular el promedio de temperatura para cada segmento de tiempo
average_temperature = df[df['PPT'].notnull()].groupby('Time_segment')['PPT'].mean()

# Crear el gráfico
plt.figure(figsize=(10, 5))  # Definir el tamaño del gráfico (opcional)

# Graficar el promedio de temperatura por segmento de tiempo con una línea y puntos
plt.plot(average_temperature.index, average_temperature.values, marker='o', linestyle='-', color='skyblue')

plt.xlabel('Segment type')  # Etiqueta del eje X
plt.ylabel('Temperature Mean')  # Etiqueta del eje Y
plt.title('Average temperature by segment type')  # Título del gráfico

plt.xticks(rotation=0)  # Rotar las etiquetas del eje X para mejor legibilidad

plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()  # Mostrar el gráfico
```
