```python
import datetime

def is_holiday(day):
    # Create a dictionary of holiday dates
    holidays = {
        '01/01/2021': 'New Year',
        '06/01/2021': 'Epiphany',
        '02/04/2021': 'Divendres Sant',
        '05/04/2021': 'Pascua',
        '01/05/2021': 'Mayo',
        '24/05/2021': 'Sant Joan1',
        '24/06/2021': 'Sant Joan',
        '11/09/2021': 'Asumpcio',
        '24/09/2021': 'Merce',
        '12/10/2021': 'Hispanidad',
        '01/11/2021': 'Tots sants',
        '06/12/2021': 'Constitucio',
        '08/12/2021': 'Constitucio2',
        '25/12/2021': 'St Esteve',
        '01/01/2022': 'New Year',
        '06/01/2022': 'Epiphany',
        '15/04/2022': 'Divendres Sant',
        '18/04/2022': 'Pascua',
        '06/06/2022': 'Pascua 2',
        '24/06/2022': 'Sant Joan',
        '15/08/2022': 'Asumpcio',
        '24/09/2022': 'Merce',
        '26/09/2022': 'Merce2',
        '12/10/2022': 'Hispanidad',
        '01/11/2022': 'Tots sants',
        '06/12/2022': 'Constitucio',
        '08/12/2022': 'Constitucio2',
        '26/12/2022': 'St Esteve',
        
    }
    # Check if the day is a holiday
    if day in holidays :
        return 'Vacation'
    else:
        return 'Not Holiday'

# Create a new column in the data frame with the holiday status
df['Holiday'] = df['Dia'].apply(is_holiday)
```

```python
def create_dies_festa_column(df):
    df['Dies_Festa'] = df['Descripcio_dia_setmana'].str.contains('Dissabte|Diumenge', case=False) | df['Holiday'].str.contains('Vacation', case=False)
    df['Dies_Festa'] = df['Dies_Festa'].replace()
    return df
```

```python
def add_ponts_column(df):
    df['Ponts'] = df['Dia'].apply(lambda x: 'Sortida Pont' if x in ['1/04/2021', '21/05/2021' '23/06/2021', '23/09/2021', '8/10/2021', '29/10/2021', '3/12/2021','5/01/2022', '14/04/2022', '3/06/2022', '23/06/2022', '12/08/2022', '23/09/2022', '28/10/2022', '2/12/2022', '23/12/2022'] else 'Res')
    return df
```

```python
def create_list_from_column(column):
    unique_strings = set(column.unique())
    clean_strings = []

    for string in unique_strings:
        if string not in clean_strings:
            clean_strings.append(string)

    return clean_strings
```
