```python
values = df['Week_day'].value_counts()
valors = values.index.values
plt.pie(values, autopct='%1.1f%%', startangle=45, radius=1)
plt.legend(valors, loc='best', title='Time segment', fontsize=3)
plt.show()
```

```python
values = df['Week_day'].value_counts()
valors = values.index.values

# Reduce the values by a factor of 50%
values = values * 0.5

plt.pie(values, autopct='%1.1f%%', startangle=45, radius=1)
plt.legend(valors, loc='best', title='Time segment', fontsize=4)
plt.savefig('pie_chart.jpg', dpi=15)
plt.show()
```
