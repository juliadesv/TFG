```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
import seaborn as sb
```

```python
df=pd.read_csv('2021_2022 meteo.xlsx - X4 diàries.csv')
```

```python
df.info()
```

```python
df
```

```python
df['TM'] = df['TM'].str.replace(',', '.').astype(float)
df['TX'] = df['TX'].str.replace(',', '.').astype(float)
df['TN'] = df['TN'].str.replace(',', '.').astype(float)
df['PM'] = df['PM'].str.replace(',', '.').astype(float)
df['PX'] = df['PX'].str.replace(',', '.').astype(float)
df['PN'] = df['PN'].str.replace(',', '.').astype(float)
df['PPT'] = df['PPT'].str.replace(',', '.').astype(float)
df['RS24h'] = df['RS24h'].str.replace(',', '.').astype(float)
df['VVM10'] = df['VVM10'].str.replace(',', '.').astype(float)
df['VVX10'] = df['VVX10'].str.replace(',', '.').astype(float)
```

```python
df['TM'] = df['TM'].astype(int)
df['TX'] = df['TX'].astype(int)
df['TN'] = df['TN'].astype(int)
df['PM'] = df['PM'].astype(int)
df['PX'] = df['PX'].astype(int)
df['PN'] = df['PN'].astype(int)
df['PPT'] = df['PPT'].astype(int)
df['RS24h'] = df['RS24h'].astype(int)
df['VVM10'] = df['VVM10'].astype(int)
df['DVVX10'] = df['DVVX10'].astype(int) 
df['VVX10'] = df['VVX10'].astype(int)
```

```python
df.info()
```

```python
df.drop('HRX', axis=1, inplace=True)
df.drop('HRN', axis=1, inplace=True)
df.drop('PX', axis=1, inplace=True)
df.drop('PN', axis=1, inplace=True)
df.drop('DVM10', axis=1, inplace=True)
df.drop('DVVX10', axis=1, inplace=True)
```

```python
df
```

```python
df1=df[['TM','TX','TN','HRM','PM','PPT','RS24h','VVM10','VVX10']]
pearsoncorr = df1.corr(method='pearson')
mask = np.triu(np.ones_like(pearsoncorr, dtype=bool))
plt.figure(figsize=(10, 8), dpi=1000)
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5,
            mask=mask)
```

```python
df1=df[['TM','TX','TN','HRM','PM','PPT','RS24h','VVM10','VVX10']]
spearman_corr = df1.corr(method='spearman')
plt.figure(figsize=(10, 8), dpi=1000)
sb.heatmap(spearman_corr, 
            xticklabels=spearman_corr.columns,
            yticklabels=spearman_corr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5,
            mask=mask)
```

```python
df.drop('TX', axis=1, inplace=True)
df.drop('TN', axis=1, inplace=True)
df.drop('VVX10', axis=1, inplace=True)
```
