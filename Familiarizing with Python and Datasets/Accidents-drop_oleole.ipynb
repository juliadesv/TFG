```python
import datetime

def is_holiday(day):
    # Create a dictionary of holiday dates
    holidays = {
        '1/01/2022': 'New Year',
        '6/01/2022': 'Epiphany',
        '15/04/2022': 'Divendres Sant',
        '18/04/2022': 'Pascua',
        '6/06/2022': 'Pascua 2',
        '24/06/2022': 'Sant Joan',
        '15/08/2022': 'Asumpcio',
        '24/09/2022': 'Merce',
        '26/09/2022': 'Merce2',
        '12/10/2022': 'Hispanidad',
        '1/11/2022': 'Tots sants',
        '6/12/2022': 'Constitucio',
        '8/12/2022': 'Constitucio2',
        '26/12/2022': 'St Esteve',
        
    }
    # Check if the day is a holiday
    if day in holidays :
        return 'Vacation'
    else:
        return 'Not Holiday'

# Create a new column in the data frame with the holiday status
df['Holiday'] = df['Dia'].apply(is_holiday)
```

```python
def create_accidents_list(df):
    accidents = []
    for index, row in df.iterrows():
        if row['Numero_expedient'] in accidents:
            accidents.append('')
        else:
            accidents.append(row['Numero_expedient'])
    return accidents
```

```python
def add_franja_horaria(df):
    df['Franja_horaria'] = df['Hora_dia'].apply(lambda x: 
                                               'Matinada' if x >= 2 and x < 6 else 
                                               'Entrada Feina' if x >= 6 and x < 10 else 
                                               'Matí' if x >= 10 and x < 14 else 
                                               'Sortida Feina' if x >= 14 and x < 18 else 
                                               'Tarda' if x >= 18 and x < 22 else 
                                               'Oci' if x >= 22 or x == 0 or x==1 or x == 2  else None)
    return df
```

```python
df_copia= add_franja_horaria(df_copia)
```

```python
def create_dies_festa_column(df):
    df['Dies_Festa'] = df['Descripcio_dia_setmana'].str.contains('Dissabte|Diumenge', case=False) | df['Holiday'].str.contains('Vacation', case=False)
    df['Dies_Festa'] = df['Dies_Festa'].replace()
    return df
```

```python
def add_ponts_column(df):
    df['Ponts'] = df['Dia'].apply(lambda x: 'Sortida Pont' if x in ['5/01/2022', '14/04/2022', '3/06/2022', '23/06/2022', '12/08/2022', '23/09/2022', '28/10/2022', '2/12/2022', '23/12/2022'] else 'Res')
    return df
```
