```python
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn import tree

# Datos de ejemplo
data = [
    ["Soleado", "Fuerte", "No"],
    ["Nublado", "Dbil", "S"],
    ["Lluvia", "Fuerte", "No"],
    ["Nublado", "Fuerte", "S"],
    ["Lluvia", "Dbil", "S"],
]

# Convertir datos categricos a numricos
le = LabelEncoder()
data_encoded = []
for i in range(len(data[0])):
    data_encoded.append(le.fit_transform([row[i] for row in data]))

# Transponer para tener caractersticas y etiquetas
X = list(zip(*data_encoded[:-1]))
y = data_encoded[-1]

# Crear y entrenar el modelo de rbol de decisin
model = DecisionTreeClassifier()
model.fit(X, y)

# Visualizar el rbol de decisin con matplotlib
plt.figure(figsize=(10, 8))
tree.plot_tree(model, feature_names=["Clima", "Viento"], class_names=["No", "S"], filled=True, rounded=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text, export_graphviz
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_iris
import graphviz
from sklearn import tree

# Datos de ejemplo
data = [
    ["Sunny", "Strong", "No"],
    ["Cloudy", "Weak", "Yes"],
    ["Rain", "Strong", "No"],
    ["Cloudy", "Strong", "Yes"],
    ["Rain", "Weak", "Yes"],
]

# Convertir datos categricos a numricos
le = LabelEncoder()
data_encoded = [le.fit_transform(column) for column in zip(*data)]

# Transponer para tener caractersticas y etiquetas
X = list(zip(*data_encoded[:-1]))
y = data_encoded[-1]

# Crear y entrenar el modelo de rbol de decisin
model = DecisionTreeClassifier()
model.fit(X, y)

# Visualizar el rbol de decisin con graphviz
dot_data = tree.export_graphviz(model, out_file=None, feature_names=["Climate", "Wind"], class_names=["No", "Yes"], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render('arbol_decision_graphviz', format='png', cleanup=True)  # Esto generar un archivo PNG llamado arbol_decision_graphviz.png
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt

# Define the features and target variable
features = [
    ["red", "large"],
    ["green", "small"],
    ["red", "small"]
]
target_variable = ["apple", "lime", "strawberry"]

# Flatten the features list for encoding
flattened_features = [item for sublist in features for item in sublist]

# Use a single LabelEncoder for all features and target variable
le = LabelEncoder()
le.fit(flattened_features + target_variable)

# Encode features and target variable
encoded_features = [le.transform(item) for item in features]
encoded_target = le.transform(target_variable)

# Create a CART classifier
clf = DecisionTreeClassifier()

# Train the classifier on the training set
clf.fit(encoded_features, encoded_target)

# Visualize the decision tree with Matplotlib
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=["Color", "Size"], class_names=le.classes_, filled=True, rounded=True)
plt.show()
```
